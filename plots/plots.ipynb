```python
import numpy as np
import matplotlib.pyplot as plt
import csv


files = ['laser_rmse.p', 'radar_rmse.p', 'fusion_rmse.p']
tiles = {'laser_rmse.p':'Laser RMSE', 'radar_rmse.p':'Radar RMSE', 'fusion_rmse.p':'Fusion RMSE'}

for file_name in files:
    file = ('/Users/matheus.souza/desenvolvimento/projetos/sdcnd/' + 
            'term2/CarND-Extended-Kalman-Filter-Project/data/' + file_name)

    pxs = []
    pys = []
    vxs = []
    vys = []

    with open(file) as tsv:
        for line in csv.reader(tsv, dialect="excel-tab"):
            if float(line[3]) < .009:
                continue
            pxs.append(line[0])
            pys.append(line[1])
            vxs.append(line[2])
            vys.append(line[3])

    x = np.arange(len(pxs))

    plt.figure(figsize=(10,5))

    plt.title(tiles[file_name])
        
    plt.ylabel('rmse')
    plt.xlabel('iterations')
    
    plt.plot(x, pxs, label="px (max:"+max(pxs)+", min:"+min(pxs)+")")
    plt.plot(x, pys, label="py (max:"+max(pys)+", min:"+min(pys)+")")
    plt.plot(x, vxs, label="vx (max:"+max(vxs)+", min:"+min(vxs)+")")
    plt.plot(x, vys, label="vy (max:"+max(vys)+", min:"+min(vys)+")")
    
    plt.grid(True)

    # Place a legend to the right of this smaller subplot.
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

    plt.show()
    

```
